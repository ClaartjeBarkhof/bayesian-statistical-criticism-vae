```python
import os
import pandas as pd

run_overview = pd.read_csv("../analysis/final_selection_runs.csv", index_col=0)
run_overview
```

```python
import shutil

dir_path = '/tmp/img'

for run_name in os.listdir("run_postprocess_files"):
    if run_name not in run_overview["run_name"].values:
        if not "mdr-vae" in run_name or "done" in run_name:
            try:
                dir_path = f"run_postprocess_files/{run_name}"
                print("Removing:", run_name)
                shutil.rmtree(dir_path)
                
            except Exception as e:
                print("Except:", e)
```
